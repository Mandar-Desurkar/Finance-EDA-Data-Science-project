```python
from pandas_datareader import data,wb
import pandas as pd
import numpy as np
import datetime
import yfinance as yf

%matplotlib inline
```

```python


start = datetime.datetime(2001, 1, 1)
end = datetime.datetime(2024, 1, 1)

BAC = pd.DataFrame(yf.download('BAC', start=start, end=end))
print("BAC Data:\n", BAC.head())

# Fetching data for Goldman Sachs
GS = pd.DataFrame(yf.download('GS', start=start, end=end))
print("Goldman Sachs (GS) Data:\n", GS.head())

# Fetching data for JPMorgan Chase
JPM = pd.DataFrame(yf.download('JPM', start=start, end=end))
print("\nJPMorgan Chase (JPM) Data:\n", JPM.head())

# Fetching data for Morgan Stanley
MS = pd.DataFrame(yf.download('MS', start=start, end=end))
print("\nMorgan Stanley (MS) Data:\n", MS.head())

# Fetching data for Wells Fargo
WFC = pd.DataFrame(yf.download('WFC', start=start, end=end))
print("\nWells Fargo (WFC) Data:\n", WFC.head())

# Fetching data for Citigroup
C = pd.DataFrame(yf.download('C', start=start, end=end))
print("\nCitigroup (C) Data:\n", C.head())
```

```python
tickers=['BAC','C','GS','JPM','MS','WFC']
```

```python
banks_stocks=pd.concat([BAC,C,GS,JPM,MS,WFC], axis=1,keys=tickers)
```

```python
banks_stocks.columns.names=['Bank Ticker', 'Stock Info']
```

```python
banks_stocks.head()
```

```python
for x in tickers:
    print(x,"",banks_stocks[x]['Close'].max())
```

```python
returns= pd.DataFrame()

for x in tickers:
    returns[x+ ""+'__return'] = banks_stocks[x]['Close'].pct_change()
returns.head()
```
